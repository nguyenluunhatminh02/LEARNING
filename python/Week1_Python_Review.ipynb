````python
# Exercise 1: mean and variance (implement without numpy)
from typing import Iterable

def mean(values: Iterable[float]) -> float:
    """Return the arithmetic mean of values."""
    values = list(values)
    if not values:
        raise ValueError("mean requires at least one value")
    return sum(values) / len(values)


def variance(values: Iterable[float]) -> float:
    """Return the population variance of values."""
    values = list(values)
    m = mean(values)
    return sum((x - m) ** 2 for x in values) / len(values)

# Test
print(mean([1,2,3,4,5]))
print(variance([1,2,3,4,5]))
````

````python
# Exercise 2: factorial recursive & iterative
def factorial_recursive(n: int) -> int:
    """Recursive factorial."""
    if n <= 1:
        return 1
    return n * factorial_recursive(n-1)


def factorial_iter(n: int) -> int:
    """Iterative factorial."""
    result = 1
    for i in range(2, n+1):
        result *= i
    return result

# Compare for n=10
print(factorial_recursive(10))
print(factorial_iter(10))
````

````python
# Exercise 3: Person class
class Person:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def birthday(self) -> None:
        """Increment the person's age by one."""
        self.age += 1

    def __repr__(self) -> str:
        return f"Person({self.name!r}, {self.age})"

# Example
p = Person('Alice', 30)
print(p)
p.birthday()
print(p)
````

````python
# Create a small sample dataset and save to sample.csv
import pandas as pd

data = pd.DataFrame({
    'id': range(1,11),
    'age': [25, 30, None, 22, 40, 35, None, 29, 31, 28],
    'income': [50000, 60000, 55000, None, 80000, 72000, 68000, None, 61000, 59000],
    'city': ['NY','LA','NY','SF','LA','NY','SF','LA','NY','SF']
})

data.to_csv('sample.csv', index=False)
print('Saved sample.csv with shape', data.shape)

data.head()
````

````python
# Data utility functions (mini `data_utils.py` inline)
from typing import Dict

def load_csv(path: str) -> pd.DataFrame:
    """Load CSV into a pandas DataFrame."""
    return pd.read_csv(path)


def summarize(df: pd.DataFrame) -> Dict:
    """Return a short summary of the DataFrame."""
    return {
        'rows': len(df),
        'columns': list(df.columns),
        'dtypes': df.dtypes.apply(lambda x: x.name).to_dict(),
        'missing': df.isnull().sum().to_dict()
    }


def drop_missing(df: pd.DataFrame, thresh: float = 0.5) -> pd.DataFrame:
    """Drop columns with more than `thresh` fraction missing. Return cleaned df."""
    if not 0 <= thresh <= 1:
        raise ValueError('thresh must be between 0 and 1')
    min_non_na = int((1 - thresh) * len(df))
    return df.dropna(axis=1, thresh=min_non_na)

# Example workflow
df = load_csv('sample.csv')
print('Summary:', summarize(df))
clean = drop_missing(df, thresh=0.5)
print('Clean shape:', clean.shape)
clean.to_csv('sample_cleaned.csv', index=False)
print('Saved sample_cleaned.csv')
````
